### Let's get prompts - System and User. 

```python
#What task they are performing and what tone they should use
system_prompt = """ 
You are a very professional writer who gives headings to the emails. 
Analyzer the emails carefully, understand the purpose and highlights of the email. And give a clear heading of no more than 20 words and no less than 8 words.
"""
```

```python
#Convo starter
user_prompt = """ 
Here is the content of email - please provide me a heading/title for this
"""
```

##### Let us get the key now  - for open ai 

```python
from dotenv import load_dotenv
import os

load_dotenv()  # loads .env from current dir or parent dirs

api_key = os.getenv("OPENAI_API_KEY")
```

### Le's make the message dict - 

```python
message = [{'role':'system' , 'content':system_prompt}, {'role':'user' , 'content':user_prompt + email}]
```

```python
message
```

### Let's get to the open AI business now 

```python
from openai import OpenAI
openai_obj = OpenAI()
```

```python
response = openai_obj.chat.completions.create(model="gpt-4.1-mini" , messages=message)
response.choices[0].message.content
```
