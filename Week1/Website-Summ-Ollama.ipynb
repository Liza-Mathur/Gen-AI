```python
import os
from dotenv import load_dotenv
from scrapper import fetch_website_contents
from IPython.display import Markdown, display
```

```python
ed = fetch_website_contents("https://edwarddonner.com")
```

```python
system_prompt = """
You are a helpful and witty assistant that analyzes website content.
You explain things clearly using simple English.
Your humor should be light, tasteful, and not distracting.
Ignore navigation menus, cookie notices, footers, and unrelated boilerplate text.
Focus only on the main informational content of the website.
Respond in markdown. Do not wrap the markdown in a code block - respond just with the markdown.
"""

developer_prompt = """
Reply in format - 
1. Key Findings - Over all summary in 5-6 key sentences in points. Not paragraph.
2. Overview - Give an overview of website - from starting to end. Overview of content.
3. Short summary - An overall short 5-6 sentences summary and contacts or qna if given in the website.
4. Your own 5 qna on website content - In Qeustion - Answer - Format.
"""

user_prompt = """
This is the website content - please proceed. +
"""
```

```python
def message_list(web):
    message = [{'role':'system', 'content':system_prompt}, {'role':'user', 'content':user_prompt + web}, {'role':'developer', 'content':developer_prompt}]
    return message
```

```python
message = message_list(ed)
```

```python
from openai import OpenAI
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
response = ollama.chat.completions.create(model="phi3", messages=message)
display(Markdown(response.choices[0].message.content))
```
