```python
import os
from openai import OpenAI
from dotenv import load_dotenv
```

```python
load_dotenv()
api_key = os.getenv('OPENAI_API_KEY')
openai_obj = OpenAI()
```

```python
OLLAMA_BASE_URL = "http://localhost:11434/v1"
ollama = OpenAI(base_url=OLLAMA_BASE_URL, api_key='ollama')
```

```python
question = """
Please explain what this code does and why:
yield from {book.get("author") for book in books if book.get("author")}
"""
System_prompt = """
You are a coding helping assisstant who either explains the code or gives output of the python or java code or both.
Understnad the question given in the user input and then understand the code, and answer the question as per the code. 
If asked about the output of the code - give clear explanation of why is this the output and what methods or methodologies are used in the code. 
"""
User_prompt = """
Here is the question and the code - 
""" + question
```

```python
message = [
    {"role":"system", "content":System_prompt},
    {"role":"user","content":User_prompt}
]
```

```python
response = ollama.chat.completions.create(model="phi3", messages=message)
print(response.choices[0].message.content)
```

```python
response = openai_obj.chat.completions.create(model="gpt-4.1-mini",messages=message)
print(response.choices[0].message.content)
```
