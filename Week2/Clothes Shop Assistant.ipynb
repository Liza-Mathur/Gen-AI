```python
system_prompt = "You are a helpful polite assistant for clothes store. \
Encourage users to buy things from here - but dont be too persuasive. \
If user asks for something that a usual cloth store doesnt have , then politely refuse for that item and promote other sales items. \
Our shoes are on 60% discount and almost everything else is on 50% discount. "
```

```python
def chat(message, history):
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role":"system", "content":system_prompt}] + history + [{"role":"user", "content":message}]
    response = openai.chat.completions.create(model = "gpt-4.1-mini", messages = messages)
    return response.choices[0].message.content
```

```python
gr.ChatInterface(fn=chat).launch()
```

```python
def chat(message, history):
    relevant_sys_prompt = system_prompt
    if 'winter' in message.lower():
        relevant_sys_prompt += "If winter items are requested for like winter jacket, winter hats or anything else winter-y, then tell them about discounts and also promote other winter items like winter wool socks, and winter innear wears and jackets (winter) - woolen too. \
        Also tell user about woolen items (meaning winter - woolen items) are at a special discount of buy 1 get 1 free if user shops for atleast 2500 ruppees"
    history = [{"role":h["role"], "content":h["content"]} for h in history]
    messages = [{"role":"system", "content":relevant_sys_prompt}] + history + [{"role":"user", "content":message}]
    response = openai.chat.completions.create(model = "gpt-4.1-mini", messages = messages)
    return response.choices[0].message.content
```

```python
gr.ChatInterface(fn=chat).launch()
```
